```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train_data=pd.read_csv("/home/shankar/Documents/train.csv")
train_data
```

```python
train_data.dropna(inplace=True)
X_train = train_data[['x']]
y_train = train_data['y']
train_data['y'].fillna(train_data['y'].mean(), inplace=True)
X_train = train_data[['x']]
y_train = train_data['y']
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)
```

```python
test_data=pd.read_csv("/home/shankar/Documents/test.csv")
test_data
```

```python
X_test = test_data[['x']]

# Predict the output (y) on the test data
y_pred = model.predict(X_test)
```

```python
print("Predicted values:")
print(y_pred)

# You can also calculate and print the mean squared error (MSE) to evaluate the model's performance
mse = mean_squared_error(test_data['y'], y_pred)
print("Mean Squared Error:", mse)
```
