```python
# Countplot for 'type' column
plt.figure(figsize=(10, 6))
sns.countplot(data=ndf, x='type')
plt.title('Count of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for 'country' column (top 10 countries)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='country', order=df['country'].value_counts().head(10).index)
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of 'release_year'
plt.figure(figsize=(12, 6))
sns.histplot(data=ndf, x='release_year', bins=30, kde=True)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for 'rating' column
plt.figure(figsize=(12, 6))
sns.countplot(data=ndf, x='rating')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for 'type' column
type_counts = ndf['type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Content Types (Movies vs. TV Shows)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Pie chart for 'rating' column
rating_counts = ndf['rating'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ratings')
plt.axis('equal')
plt.show()
```

```python

# Pie chart for 'type' column using Seaborn
type_counts = ndf['type'].value_counts()
plt.figure(figsize=(8, 8))
sns.set_palette('pastel')  # Set a pastel color palette
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Content Types (Movies vs. TV Shows)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Pie chart for 'rating' column using Seaborn
rating_counts = ndf['rating'].value_counts()
plt.figure(figsize=(10, 8))
sns.set_palette('Set3')  # Set a different color palette
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ratings')
plt.axis('equal')
plt.show()
```
